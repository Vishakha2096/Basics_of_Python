```python
##Write a Python program to calculate the factorial of a number provided by the user using a while loop.
```

```python
num = int(input("2: "))
if num < 0:
    print("not defined .")
else:
    factorial = 1
    i = 1
    while i <= num:
        factorial *= i
        i += 1

    print(f"Factorial of {num} is {factorial}")
```
